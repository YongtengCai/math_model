## 特征识别

```python
#提取loc
import pandas as pd

# 导入data
data = pd.read_csv('result_table.csv')
data
```

```python
# 定义函数来解析loc列的字符串
def parse_loc_column(loc_value):
    # 检查loc_value是否为字符串类型
    if not isinstance(loc_value, str):
        return None, None
    
    # 按下划线分割字符串
    parts = loc_value.split('_')
    
    # 检查是否至少有3个部分
    if len(parts) < 3:
        return None, None
    
    # 提取第一部分作为frq
    frq = parts[0]
    
    # 提取第二部分作为新的loc
    new_loc = parts[1]
    
    return frq, new_loc

# 确保data中有loc列
if 'loc' in data.columns:
    # 应用函数到每一行数据
    parsed_results = data['loc'].apply(parse_loc_column)
    
    # 将结果拆分为两列
    data['frq'] = parsed_results.apply(lambda x: x[0])
    data['loc'] = parsed_results.apply(lambda x: x[1])
    
    # 可选：如果需要用新的loc替换原来的loc列，可以使用以下代码
    # data['loc'] = data['new_loc']
    # data.drop('new_loc', axis=1, inplace=True)

data
```
